# Using Python in the Cloud!  

Python is a popular choice for DevOps! It is quick to write,
we can easily make small little scripts or full blown applications
with this tool. So you can see why it has a popular following in the community.  

There are many Cloud providers at the moment both public and Private and all
of them will generally offer and API for Developers to (ab)use!  

Even if Python is not being used to directly orchestrate a Cloud Stack, there
are many, many examples where Python projects are used to create the DSL
(Domain-specific language) that powers the Cloud Stack, an example of this is the
`Troposphere` (See [here](https://github.com/cloudtools/troposphere)) Project for
generating [AWS CloudFormation](https://aws.amazon.com/cloudformation/).  

Let's take a look at each Cloud in turn!
