# Using Python to make Web Sites!

No matter if we are looking for a small piece of `html`
code or a full blow Web Application. Python has a tool to help
us out to master the craft of Web Development.

Some companies exclusively use certain Web Frameworks and languages
where as others chop and change depending on their clients needs.

In more recent times there have been "better" (matter of opinion really!)
languages to code in for use on the Web. Facebook still use PHP to create
their website where as Uber will use Node.js and Go heavily.

Saying that, large companies like Netflix are known to use Python heavily
at the time of writing this Tutorial.

This guide is going to look at some of the Web Frameworks available in Python
for us to play with. It would be good to already know some HTML, CSS and JavaScript
as these will help us learn how to make some really pretty Websites but are beyond
the scope of these tutorials.

You can always use [W3Schools](https://www.w3schools.com/) and [Code Acadamy](https://www.codecademy.com/) as a guide if you want to learn
more or understand something I have used in my code better!

For the purposes of this guide all we need to really know it that:
1. HTML dictates the content of our Web Site
2. CSS dictates the style (colours, background, etc)
3. JavaScript (or it's many Frame works!) provide a programming language to run on the users computer (Great for Animation... or attack!)
