# Using Python to create Games Big and Small!

Python is a very versatile language that can be used in many
settings. The games industry has caught on to this, we have
seen how easy it is to create simple logic puzzles on the
command line and inside the terminal. But how about making
more complex games? Where are the pretty graphics?

The community has created many, many Modules for us to use
in this specialty we are going to take a look at some of them
and there different capabilities. It should give you as the
Developer an idea of what you like to use, to create what you
want to create!

The only limitation should be your imaginations at this point!
