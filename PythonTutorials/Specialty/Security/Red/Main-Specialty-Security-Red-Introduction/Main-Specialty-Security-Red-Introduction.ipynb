## Python as a PenTest tool!
Anyone who is interested in Security (especially the attacking side)
will know the tool [Metasploit](https://www.metasploit.com/) which is a
beautiful attack platform that comes with a nice interface to wrap our attacks
in and make the job a lot easier... Sadly, (opinion) this framework is
currently written in Ruby. However, in the past it was originally written in
Python and will even (to this day of writing) will run Python code as part
of the exploit process!

Other tools and uses that the language has is automating the boring tasks.
Some parts of being evil are so very, very boring! Hash cracking,
Brute-Forcing Passwords or Fuzzing Applications come to mind for a few.

Of course part of helping the above is working smart to cut down the number
of tries needed to get the desired result, understanding a problem and
removing several guessing can save days!
