# Healing Python  

As well as Attack, Python can also be useful to help prevent attacks. It has
a large range of tools to help build automation to detect and fix potential
flaws in the system.  

It also allows us as Security Professionals to quickly craft PoC (Proof of
Concept) scripts to quickly test different variations of attacks against our
own systems and make sure that they are not exposed for APTs (Advanced
Persistent Threats) to break in to our Organisations and use the Network for
their needs.  

Below are a few exercises that illustrate the different uses of Python to give
you more exposure to the language and how it can be used.
