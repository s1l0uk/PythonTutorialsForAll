# Python on Lock Down!

Python (like most awesome Technology!) has a darker side. It has
been known to be used for the naughtier tasks with in the industry.
In this case, Computer Security.  

With this tool we can test computer's security (Red Team - Attack!) as well as
look to remediate issues with the language (Blue Team - Defending).Sometimes
the usage of Python is relegated to preparing data. In the fast pace world of
Security this can be very useful as it gives us speed to react to any given
scenario. An example of this is how we can run a playbook and automate it with
Python to shut down malicious activity. Alternatively, we can use Python to
automate the more boring aspects of the preparation needed to hack into a
computer.  

Below are a collection of Security based exercises... Choose your team!
