# Python Specialties

Python is a great language (As we have seen)!  

Simple, elegant and very quick to pick up and play with. As mentioned at the
start of the course, knowledge of the language (or any language) comes with
exposure.  

We learn how to code by seeing others code and playing with new modules that
fixes a problem we are experiencing. In an Attempt to show some more cases
where Python can be used and allow you to get a taste of the different specialties
we have created a collection for you to work with. Everything from Security, to
Gaming and even some DevOps!
