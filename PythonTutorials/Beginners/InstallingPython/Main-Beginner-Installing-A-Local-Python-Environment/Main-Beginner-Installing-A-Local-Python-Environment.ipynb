# Installing a Python Environment

While the `ipy` notebooks provided by Google are a great way to learn
the Python language and syntax. Some of the exercises (especially the
more in depth and visual ones) will require the use of your local
computer.

The problem with this exercise is there is so much variation possible
and it would be wrong to force you to use a paradigm. So below are
some tutorials to help you get set up with a Python Environment you can
run locally.
