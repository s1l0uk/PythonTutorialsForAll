this could be a fun exercise! We will probably need to install it.

```
python3
Python 3.6.9 (default, Nov  7 2019, 10:44:02)
[GCC 8.3.0] on Linux
Type "help", "copyright", "credits" or "license" for more information.
>>>  
```

```
>>> print("😹")
😹
```  

## Let's get started!  

Ok! So you either got an error or you don't have it installed.
Let's walk through the Installation Process and see if we can
make it work for you.  

Before we go any further, if you are installing this do your local
computer you will need to have root access and we do not recommend
installing packages to your company computer. We also cannot take
responsibility for any problems the installations may cause to your
computer (other applications).  

These guides are provided in good faith and are based of the Official
instructions from the Python Foundation on their website in a simplified
format to help the curious adopt the language.  

### Get the latest Python3 from Python Foundation  
Firstly, we will need to get a windows installer which can be found
[here](https://www.python.org/ftp/python/3.8.2/). Which will provide
us with the version 3.8.2 (If we need another we can change the URL).

If we are running a 32 bit version of windows we will require the 
`.msi` file we can add `/win32/` to the URL to get our required version.

Once we have located what we need, we just click the link to download:  
[version 3.8.2 .exe](https://www.python.org/ftp/python/3.8.2/win32/core.msi)
[version 3.8.2 .msi](https://www.python.org/ftp/python/3.8.2/python-3.8.2rc2.exe)

Once we have the Installer let's move on to next phase!  

### Run the Installer  
Running the installer on Windows is simple, we locate the package we downloaded
above and double click the icon.  

We can also run this from the command line to get used to working with it.
However, they both produce the same results - The Install Window!  

Note: We can make the installer run automatically which is useful on headless
systems but going through the install process gives you more control and takes
you through what is happening.  

### Add Python3 to the PATH  
Sneakily the Installer puts small option down the bottom to ensure that the
product we are installing is added to our `PATH`. This will be required to
make using Python easier in the Windows environment.  

#### The PATH
The `PATH` variable is avilable on all Operating Systems. This variable is
a list that provides the system an order to search itself for programs that
we may want to run.  

By ticking the box above we tell Windows that we want to be able to specify
`python` on the command line and have this program run for us.  

Once the box has been ticked, Let's move on!  

###Customize the install  
Let's drill down into our install, Click `Customize Instalation`
CustomiseInstall.png

This brings up features of what you may or may not wish to install (if your
computer does not have much space you might want to opt out some of the files).

In the first instance you can ignore `Python test suite` and `IDLE` development
environment. You may want to bring your own to the party later...

It is recommened to select the option for installing for all users as this can
make developing easier in the future but might not suit your needs.

Most of the Defaults should suit your needs so it is OK to click through
the options and finally the `Install` button to confirm your selections and
add Python to your system.

###Finishing The Installation
Because Windows has long filenames there is an option included on most installs
to `disable path limit length`. In Development environments it is recommended to
select this option as it will, again, ease development for ourselves in the long run.

Note: You will be prompted to allow trust to the Application to make changes to the Windows
Operating system.

## Wrap Up!

Hopefully now you have a working set up of Python. This is the end of this
part of the Tutorial. Creating a more advanced Setups will become available
as guides. However, they can also get opinionated to the best way to manage
the Environment as the community has developed many different tools to
make developing in Python in different scenarios much easier!
