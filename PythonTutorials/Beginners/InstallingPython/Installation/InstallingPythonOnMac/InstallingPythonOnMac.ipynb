# Installing Python on Mac

Python generally comes installed by default on Mac systems so
this should be a rather simple exercise. We may need to install
XCode (or at least accept the License especially if we want to 
use `git`).  

To test this out we can issue the following command if we open up a
terminal:  
```
$ python
```
or  
```
$ python3
Python 3.6.9 (default, Nov  7 2019, 10:44:02)
[GCC 8.3.0] on Linux
Type "help", "copyright", "credits" or "license" for more information.
>>>  
```
if this works we should be able to type:  
```
>>> print("😹")
😹
```  

## But it didn't work!  

Ok! So you either got an error or you don't have it installed.
Let's walk through the Installation Process and see if we can
make it work for you.  

Before we go any further, if you are installing this do your local
computer you will need to have root access and we do not recommend
installing packages to your company computer. We also cannot take
responsibility for any problems the installations may cause to your
computer (other applications).  

These guides are provided in good faith and are based of the Official
instructions from the Python Foundation on their website in a simplified
format to help the curious adopt the language.  

### The AppStore  
In some cases the Apple AppStore will contain a copy of the latest
official Python for you to click and install. If you have configured your
computer around the AppStore we would recommend continuing this way if you
can.  

At the very least, if you are beginning your development career from this
Computer, we recommend installing the XCode packages either through the
AppStore or using the Terminal and entering:  
```
xcode-select --install
```
to install the xcode command line tools and:  
```
xcodebuild -license
```
Which will present you with a License to read and accept... The Price for
using Apple...  

### The Brew Installer  
Some people will use `brew` to install packages to their Computer, the
website can be found [here](https://brew.sh/).  

It can be installed with:
```
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install.sh)"
```
but will require the XCode setup mentioned above!  

Once we have brew installed we can install Python with:  
```
brew install python3
```

### The WebSite - pkg  
Finally if all else fails, the Python Foundation provide an FTP server
with many versions of Python to be installed [here](https://www.python.org/ftp/python/).  

We can drill down and get a specified version by adding the number to the URL:  
```
https://www.python.org/ftp/python/3.8.2/
```
and from the list we only need to pick the version we need that contains our
Operating system, in this case: `macosx`  

Once we find our installer we can download and point and click to install Python,
for our purposes the default settings should work for us.  


````
# Installing Mac is THIS simple!
!/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install.sh)"
!alias PACKAGE_MANAGER="brew"
!PACKAGE_MANAGER install python3
````
