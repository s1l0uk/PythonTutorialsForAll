# Putting the OOP in OOPS

In these Chapters we are going to explore the world
of Object Orientated Programming with a Python flavor.

OOP Programming can be of great use to us as developers and enables
us to do more with the code, including write less of it!

It can be a difficult to get the concept, but much like riding a bike,
(anyone who has read the Oracle guide will get the pun) once you get it,
you will understand it for Life!

The fun thing about OOP is that not a lot of professionals understand it,
let alone can code it (...well!). The reason being that there is so much
variation and the "Official Guides" can be difficult to read with all the
Jargon.

Hopefully these chapters will simplify the subject and spark your imagination
to realise the different use cases for this way of programming.
